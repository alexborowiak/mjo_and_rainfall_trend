```python
ROOT_DIR = '/g/data/zv2/agcd/v1/precip/calib/r005/01day/'
```

```python
os.listdir(ROOT_DIR)
```

```python
awap_2 = xr.open_mfdataset(ROOT_DIR + '*.nc', chunks = {'lat':100,'lon':100,'time':-1})
```

```python
awap_2
```

```python
np.unique(awap_2.time.dt.year.values)
```

```python
awap_sub = awap_2.where(awap_2.time.dt.year.isin(np.arange(1970,2021)), drop = True).precip
awap_sub = awap_sub.where(awap_sub.time.dt.month.isin([10,11,12,1,2,3]))
awap_sub = awap_sub.where(np.logical_and(awap_sub >1,
                                        awap_sub < 8000))
```

```python
awap_sub
```

```python
np.unique(awap_sub.time.dt.year)
```
